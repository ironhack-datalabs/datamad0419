```python
# Normalizo los valores de la columna offices, creando un dataframe a parte con sus valores separados en columnas.

mydataset['offices2'] = mydataset.offices.apply(lambda x: x[0])
locations = json_normalize(mydataset['offices2'])
display(locations.head())
print(locations.shape)
```

```python
# Para crear la columna de coordenadas, creo una lista para juntar lat y long, para después convertirla en columna.

coordinates = []

for i in range(len(locations["latitude"])):
    coordinates.append([locations["latitude"][i], locations["longitude"][i]])
    
locations["coordinates"] = coordinates
locations.head()
```

```python
# Uno el dataframe de oficinas a mi dataframe original

mydataset = mydataset.join(locations)
mydataset.head()
```

```python
# Y elimino las columnas que ya no me sirven.

mydataset = mydataset.drop(["offices", "offices2"], axis = 1)
mydataset.head()
```

```python
# De la columna funding_rounds, me quedo con el total dinero levantado y la moneda, para normalizar el cambio después.

raised_amount = []

for lista in mydataset["funding_rounds"]:
    amount = 0
    for dic in lista:
        amount += dic.get("raised_amount")
    raised_amount.append(amount)

raised_curr_code = []

for lista in mydataset["funding_rounds"]:
    raised_curr_code.append(lista[0].get("raised_currency_code"))
```

```python
# Creo columnas a partir de las listas raised_amount y raised_curr_code:

mydataset["raised_amount"] = raised_amount
mydataset["raised_amount_currency"] = raised_curr_code
mydataset = mydataset.drop("funding_rounds", axis = 1)

display(mydataset.head())
```

```python
# Me doy cuenta de que la columna total_money_raised equivale a mi nueva columna raised_amount...

mydataset = mydataset.drop(["raised_amount", "raised_amount_currency"], axis = 1)

mydataset.head()
```
