```python
# Miro a ver los duplicados de cada columna

def duplicates(data):
    res = {}
    print("Dataset rows:", len(data.index))
    for col in data.columns:
        values = []
        for v in data["{}".format(col)]:
            values.append(v)
        res.update({col: len(set(values))})
    return res


print(duplicates(data))
```
