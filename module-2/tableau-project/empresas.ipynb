Señas de nuestra empresa 
Nos dedicamos a los videojuegos y nos componemos de unos 50 empleados. 
Queremos buscar una nueva oficina sabiendo que:
-Contamos con 20 desarrolladores, 20 UX/UI y 10 ejecutivos.
Y queremos además:
    - Existan compañías de Software cerca.
    - Buen ratio entre grandes compañías y start-ups
    - Intentar que existan empresas que pueden ayudar a nuestros empleados(comercios, ¿Restaurantes?)
    - Intentar que no existan compañias con muchos años 

Con este ejercicio se quiere verificar si simplemente con una aproximanción visual y representativa de los datos es suficiente o hace falta una mayor comprensión de los datos para poder encontrar el mejor punto

```python
from pymongo import MongoClient
client = MongoClient()
db = client.companies
import pandas as pd
import re
```

```python
bla = db.companies.find({ "$and": [ { "offices.latitude": {"$exists": True,"$ne":None} },{ "offices.longitude": {"$exists": True,"$ne":None} }, {"founded_year" : {"$gte": 2000} }, {'deadpooled_year': {"$eq":None}},{'number_of_employees':{'$ne':None}}]},{'name':1,'_id':0,'founded_year': 1, 'number_of_employees':1,'offices.latitude':1, 'offices.longitude':1,'category_code':1,'total_money_raised':1})  
df = pd.DataFrame(bla)
```

```python
#crear columna latitud
def latitud(coord): 
    return coord[0]['latitude']

#crear columna longitud   
def longitud(coord): 
    return coord[0]['longitude']

# crear nuevas columnas en el DataFrame 
def insert_latlong(df):   
    df['latitude'] = df['offices'].apply(latitud)
    df['longitude'] = df['offices'].apply(longitud)
    return df
```

```python
df= insert_latlong(df)
df = df.drop(['offices'], axis=1)
df
```

```python
def geoj(a,b):
    return {'type':'Point', 'coordinates':[a,b]}

df['geojson'] = df.apply(lambda x: geoj(x['longitude'], x['latitude']),axis=1)
df
```

```python
df.to_json("dflimpia.json",orient="records",lines=True)
```
