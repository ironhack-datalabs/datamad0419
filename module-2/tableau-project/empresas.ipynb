```python
from pymongo import MongoClient
client = MongoClient()
db = client.companies
import pandas as pd
import re
```

```python
total1 = db.companies.find({ "$and": [ { "offices.latitude": {"$exists": True,"$ne":None} },{ "offices.longitude": {"$exists": True,"$ne":None} }, {"founded_year" : {"$gte": 2000} }, {'deadpooled_year': {"$eq":None}},{'number_of_employees':{'$ne':None}},]},{'name':1,'_id':0,'founded_year': 1, 'number_of_employees':1,'offices.latitude':1, 'offices.longitude':1,'category_code':1,'total_money_raised':1})  
grande1 = db.companies.find({ "$and": [ { "offices.latitude": {"$exists": True,"$ne":None} },{ "offices.longitude": {"$exists": True,"$ne":None} }, {"founded_year" : {"$gte": 2000} }, {'deadpooled_year': {"$eq":None}},{'number_of_employees':{"$gte": 50}},{'$or':[{'category_code': 'web' },{'category_code': 'software' },{'category_code': 'games_video' },{'category_code': 'mobile' },{'category_code': 'music' },{'category_code': 'nanotech'}]}]},{'name':1,'_id':0,'founded_year': 1, 'number_of_employees':1,'offices.latitude':1, 'offices.longitude':1,'category_code':1,'total_money_raised':1})  
startup1 = db.companies.find({ "$and": [ { "offices.latitude": {"$exists": True,"$ne":None} },{ "offices.longitude": {"$exists": True,"$ne":None} }, {"founded_year" : {"$gte": 2000} }, {'deadpooled_year': {"$eq":None}},{'number_of_employees':{'$lt':50,'$gt':1}},{'$or':[{'category_code': 'web' },{'category_code': 'software' },{'category_code': 'games_video' },{'category_code': 'mobile' },{'category_code': 'music' },{'category_code': 'nanotech'}]}]},{'name':1,'_id':0,'founded_year': 1, 'number_of_employees':1,'offices.latitude':1, 'offices.longitude':1,'category_code':1,'total_money_raised':1})  
servicios1 = db.companies.find({ "$and": [ { "offices.latitude": {"$exists": True,"$ne":None} },{ "offices.longitude": {"$exists": True,"$ne":None} }, {'deadpooled_year': {"$eq":None}},{'$or':[{'category_code': 'hospitality' },{'category_code': 'medical' },{'category_code': 'news' },{'category_code': 'education' },{'category_code': 'music' }]}]},{'name':1,'_id':0,'founded_year': 1, 'number_of_employees':1,'offices.latitude':1, 'offices.longitude':1,'category_code':1,'total_money_raised':1})  

total = pd.DataFrame(total1)
grande = pd.DataFrame(grande1)
startup = pd.DataFrame(startup1)
servicios = pd.DataFrame(servicios1)
```

```python
#crear columna latitud
def latitud(coord): 
    return coord[0]['latitude']

#crear columna longitud   
def longitud(coord): 
    return coord[0]['longitude']

# crear nuevas columnas en el DataFrame 
def insert_latlong(df):   
    df['latitude'] = df['offices'].apply(latitud)
    df['longitude'] = df['offices'].apply(longitud)
    return df
```

```python
#creamos las columnas que necesitamos en las bases de datos de empresas grandes y pequeñas y en el total
total= insert_latlong(total)
grande= insert_latlong(grande)
startup= insert_latlong(startup)
servicios= insert_latlong(servicios)
total = total.drop(['offices'], axis=1)
grande = grande.drop(['offices'], axis=1)
startup = startup.drop(['offices'], axis=1)
servicios = servicios.drop(['offices'], axis=1)
total
```

```python
#creamos la columna geojson para posteriormente en pymongo hacer el indice
def geoj(a,b):
    return {'type':'Point', 'coordinates':[a,b]}

total['geojson'] = total.apply(lambda x: geoj(x['longitude'], x['latitude']),axis=1)
total
```

```python
#ahora se dibujan los puntos interesantes según sean 
import folium
import pandas as pd


# Make an empty map
m = folium.Map(location=[20, 0], tiles="Mapbox Bright", zoom_start=2)

# I can add marker one by one on the map
for i in range(0,len(grande)):
    folium.Marker([grande.iloc[i]['latitude'], grande.iloc[i]['longitude']], popup=grande.iloc[i]['name'],icon=folium.Icon(color='green')).add_to(m)
for i in range(0,len(startup)):
    folium.Marker([startup.iloc[i]['latitude'], startup.iloc[i]['longitude']], popup=startup.iloc[i]['name'],icon=folium.Icon(color='blue')).add_to(m)
for i in range(0,len(servicios)):
    folium.Marker([servicios.iloc[i]['latitude'], servicios.iloc[i]['longitude']], popup=servicios.iloc[i]['name'],icon=folium.Icon(color='red')).add_to(m)


# Save it as html
#m.save('312_markers_on_folium_map1.html')
```
