```python
datos_cercania=[]
for i in range(len(df["latitude"])):
        empresa2 = pd.DataFrame(db.empresas.find({
        "geojson": {
         "$near": {
           "$geometry": {
              "type": "Point" ,
              "coordinates": [ df["longitude"][i] , df["latitude"][i] ]
           },
           "$maxDistance": 1000, # In meters
             }
           }
        }))
        datos_cercania.append(len(empresa2))

        
```

```python
df["empresas_cerca"]=datos_cercania
df.sort_values(["empresas_cerca"], ascending=[False], inplace=True) 
df
```
